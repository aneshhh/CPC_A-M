```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.svm import SVC, SVR
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

# Step 1: Upload training and testing files
uploaded = files.upload()

# Load training and testing data
train_data = pd.read_csv("/content/assignment_files-1.csv")  # Training set
test_data = pd.read_csv("/content/assignment_files-2.csv")   # Testing set

# EDA on Training Data
print("Training Data Overview:")
print(train_data.head())

# Check for missing values
print("\nMissing Values in Training Data:")
print(train_data.isnull().sum())

# Data types and summary statistics
print("\nData Information:")
print(train_data.info())
print("\nSummary Statistics:")
print(train_data.describe())

# Handle missing values for EDA (temporary for visualization purposes)
for col in train_data.columns:
    if train_data[col].dtype == "object":
        train_data[col].fillna(train_data[col].mode()[0], inplace=True)
    else:
        train_data[col].fillna(train_data[col].mean(), inplace=True)

# Visualize data distribution: Histograms for numeric features
numeric_features = train_data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 10))
numeric_features.hist(bins=20, color="teal", edgecolor="black", figsize=(12, 10))
plt.suptitle("Histograms of Numeric Features (Training Set)", fontsize=16)
plt.tight_layout()
plt.show()

# Correlation heatmap for numeric features
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_features.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features (Training Set)", fontsize=16)
plt.show()

# Box plots to analyze distributions and outliers in numeric features
plt.figure(figsize=(14, 8))
for i, col in enumerate(numeric_features.columns):
    plt.subplot((len(numeric_features.columns) + 2) // 3, 3, i + 1)
    sns.boxplot(y=train_data[col], color="skyblue")
    plt.title(f"Box Plot: {col}")
plt.tight_layout()
plt.show()

# Scatter plots: Relationships between key numeric features and target variable
# Replace 'LoanEligibility' with the actual target column name if different
if "LoanEligibility" in train_data.columns:
    for col in numeric_features.columns:
        if col != "LoanEligibility":
            plt.figure(figsize=(8, 6))
            sns.scatterplot(x=train_data[col], y=train_data["LoanEligibility"], alpha=0.7)
            plt.title(f"{col} vs LoanEligibility Scatter Plot")
            plt.xlabel(col)
            plt.ylabel("LoanEligibility")
            plt.show()

# Highlight Key Features Influencing Loan Eligibility
# Correlation with the target variable
if "LoanEligibility" in train_data.columns:
    corr_with_target = train_data.corr()["LoanEligibility"].sort_values(ascending=False)
    print("\nFeatures Highly Correlated with Loan Eligibility:")
    print(corr_with_target)
else:
    print("LoanEligibility column not found for correlation analysis.")



#ML Part




# Handle missing values for classification task
train_data.dropna(subset=['Loan_Status'], inplace=True)
train_data['LoanAmount'].fillna(train_data['LoanAmount'].mean(), inplace=True)
train_data['Loan_Amount_Term'].fillna(train_data['Loan_Amount_Term'].mode()[0], inplace=True)

# Encode Loan_Status
train_data['Loan_Status'] = train_data['Loan_Status'].map({'Y': 1, 'N': 0})

# Separate features and target for classification
X_class = train_data.drop(['Loan_ID', 'Loan_Status'], axis=1)
y_class = train_data['Loan_Status']

# Fill missing values and encode categorical data for classification
for col in X_class.columns:
    if X_class[col].dtype == "object":
        X_class[col].fillna(X_class[col].mode()[0], inplace=True)
        test_data[col].fillna(test_data[col].mode()[0], inplace=True)
    else:
        X_class[col].fillna(X_class[col].mean(), inplace=True)
        test_data[col].fillna(test_data[col].mean(), inplace=True)

X_class = pd.get_dummies(X_class, drop_first=True)
test_data = pd.get_dummies(test_data, drop_first=True)

# Align train and test data
X_class, test_data = X_class.align(test_data, join='inner', axis=1)

# Split data for classification
X_train, X_val, y_train, y_val = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Standardize features
scaler_class = StandardScaler()
X_train_scaled = scaler_class.fit_transform(X_train)
X_val_scaled = scaler_class.transform(X_val)
test_data_scaled = scaler_class.transform(test_data)

# Classification Models
classification_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(probability=True)
}

# Fit and evaluate classification models
for model_name, model in classification_models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_val_scaled)
    y_proba = model.predict_proba(X_val_scaled)[:, 1]

    print(f"\n{model_name} Classification Performance:")
    print(f"Accuracy: {accuracy_score(y_val, y_pred):.2f}")
    print(f"Precision: {precision_score(y_val, y_pred):.2f}")
    print(f"Recall: {recall_score(y_val, y_pred):.2f}")
    print(f"F1 Score: {f1_score(y_val, y_pred):.2f}")
    print(f"ROC AUC: {roc_auc_score(y_val, y_proba):.2f}")

    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_val, y_proba)
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc_score(y_val, y_proba):.2f})")

plt.plot([0, 1], [0, 1], linestyle='--')
plt.title("ROC Curves for Classification Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# Hyperparameter tuning for Random Forest (Classification)
param_grid_class = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20, 30]}
grid_search_class = GridSearchCV(RandomForestClassifier(), param_grid_class, cv=5, scoring='accuracy')
grid_search_class.fit(X_train_scaled, y_train)
print("\nBest Parameters for Random Forest (Classification):", grid_search_class.best_params_)

# Regression Task for LoanAmount and Loan_Amount_Term
X_reg = train_data.drop(['Loan_ID', 'Loan_Status', 'LoanAmount', 'Loan_Amount_Term'], axis=1)
y_max_loan = train_data['LoanAmount']
y_min_duration = train_data['Loan_Amount_Term']

# Encode categorical features for regression
X_reg = pd.get_dummies(X_reg, drop_first=True)

# Align with test data
X_reg, _ = X_reg.align(test_data, join='inner', axis=1)

# Standardize regression features
scaler_reg = StandardScaler()
X_reg_scaled = scaler_reg.fit_transform(X_reg)

# Regression models
reg_models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Gradient Boosting Regressor": GradientBoostingRegressor(),
    "SVM Regressor": SVR()
}

# Fit regression models and evaluate for LoanAmount
for model_name, model in reg_models.items():
    model.fit(X_reg_scaled, y_max_loan)
    y_max_pred = model.predict(X_reg_scaled)
    print(y_max_pred)

    print(f"\n{model_name} Regression (Maximum Loan) Performance:")
    print(f"R: {r2_score(y_max_loan, y_max_pred):.2f}")
    print(f"MAE: {mean_absolute_error(y_max_loan, y_max_pred):.2f}")
    print(f"MSE: {mean_squared_error(y_max_loan, y_max_pred):.2f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_max_loan, y_max_pred)):.2f}")

    # Plot residuals
    plt.figure(figsize=(8, 6))
    sns.residplot(x=y_max_loan, y=y_max_pred, lowess=True, color="g")
    plt.title(f"{model_name} Residual Plot for Maximum Loan Prediction")
    plt.show()

    # Plot actual vs predicted
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_max_loan, y=y_max_pred, alpha=0.7)
    plt.title(f"{model_name} Actual vs Predicted for Maximum Loan")
    plt.xlabel("Actual Loan Amount")
    plt.ylabel("Predicted Loan Amount")
    plt.show()




```
